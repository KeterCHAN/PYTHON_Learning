## Mirror and copy

```python
lista = [1, 2, 4, 6, 8]
listb = lista
# this is mirroring.
# if one changes, the other will also change
```

```python
listb[1] = -1000
print(listb)
print(lista)
# change listb, lista will change
```

```python
# change lista, listb will also change
del lista[-1]
```

```python
lista
```

```python
listb
```

```python
# in order to make a copy without mirroring
# use copy() method

listc = [2, 3, 0, 9]
listd = listc.copy()
listd.remove(2)
listd
```

## `join` substrings to a string

```python
# join tuples/lists
num_tuple = ('1', '2', '3', '4')
separator = '->-'
print(separator.join(num_tuple))
```

```python
# join dictionaries
aDict = {'mat': 1, 'that': 2}
sep = '^'

# joins the keys only
print(sep.join(aDict))
```

# Dictionaries

## Define a dictionary

```python
grades = {'Chao' : 95, 'Eric' : 82}
numbers = {2: "Two", 3: "Three", 2: "Too"}
# dictionary keys must be unique
# duplicates are overriden
eng2can = dict()
# creates an empty dictionary
print(eng2can)
print(numbers)
```

```python
grades['Chao']
```

```python
# merge two dictionaries
{**grades, **numbers}
```

```python
grades | numbers
```

```python
dict(grades | numbers)
```

```python
# add key-value pairs to a dictionary
eng2can['one'] = 'yat'
eng2can['two'] = 'yi'
eng2can['three'] = 'saam'
print(eng2can)
```

```python
'four' in eng2can
```

## Use dictionary as a counter

```python
print(grades.get('Chao', 59))
print(grades.get('Chuck', 59))
```

```python
# to count letter frequency
word = 'Supercallifragilisticexpialidocious'
count = dict()
# setup an empty dictionary as a counter
for c in word:
    count[c] = count.get(c, 0) + 1
print(count)
```

```python
# alternatively, use the Counter module

from collections import Counter

freq = Counter(word)
```

```python
freq
```

```python
freq.most_common(5)
```

## Traverse a dictionary elements

```python
print(eng2can.keys())
print(eng2can.values())
print(eng2can.items())
```

```python
# turn the sequence to a list or a tuple
print(tuple(eng2can.keys()))
print(list(eng2can.items()))
```

```python
# Traverse the keys
for key in eng2can:
    print(f"{key.upper()} in Cantonese is '{eng2can[key]}'")
```

```python
# another way to traverse the keys
for key in sorted(eng2can.keys(), reverse = True):
    print(f"{key.upper()} in Cantonese is '{eng2can[key]}'")
```

```python
# two iteration variables
for eng, can in eng2can.items() :
    print(f"{eng.upper()} in Cantonese is '{can}'")
```

## Complex dictionaries

```python
departments = {
                'business': ['accounting','finance','economics'],
                'linguistics': ['chinese','english','korean']
            }
departments['linguistics'][0]
```

```python
for college, majors in departments.items():
    print(college.upper() + ':')
    for i in majors:
        print(' ', i.capitalize())
```

```python
people = {
            1: {'Name': 'John', 'Age': '27', 'Sex': 'Male'},
            2: {'Name': 'Marie', 'Age': '22', 'Sex': 'Female'}
        }

for name in people.values(): print(name['Name'])
```
