# `for` loop

## Construct simple for loops

```python
for x in [5, 4, 3, 2, 1]:
    print(x)
print('The End!')
```

```python
for letter in "HKU Business School":
    # a text is a sequence of letters
    # it's also considered an iterable
    print(letter, end = "-")
```

```python
# use a for loop to count frequency and take sum

count = 0
total = 0
# set the initial values

for item in [9, 41, 12, 3]:
    count += 1
    total += item
    print('Value =', item, '\tCount =', count, '\tCurrent total =', total)
print('Final count is', count)
print('Final total is', total)
```

```python
# if you dont know how to setup the initial value, then set it to be None.
# then replace the initial value in the loop.

total = None

for item in [9, 41, 12, 3]:
    if total == None:
        # may also use keyword: is
        total = item
    else:
        total += item
    print('Value =', item, '\tTotal =', total)
print('Final total is', total)
```

```python
# use if statement in loops
for y in range(6):
    if y % 2 == 0:
        print(y, 'is an even number.')
    else:
        print(y, 'is an odd number.')
```

```python
# a one liner. It's not recommended for beginners
["Even" if y % 2 == 0 else "Odd" for y in range(6)]
```

## Nested `for` loops

```python
for x in range(3):
    for y in range(4):
        print(y)
    print()
    # this last print produces an empty line
```

```python
for x in range(3):
    for y in range(4):
        print(y, end = ' ')
    print()
    # this last print produces an empty line
```

```python
for x in range(5):
    for _ in range(x + 1):
        # we use _ as the interation variable
        # we do not need to use its value
        print('*', end = '')
    print()
```
