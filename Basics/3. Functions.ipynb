## Define your own functions

```python
def greet(lang): # adding a parameter
    if lang == 'jp':
        print('Konichiwa')
    elif lang == 'fr':
        print('Bonjour')
    elif lang == 'thai':
        print('Sawadika')
    else:
        print('Hello')

greet('JP')
```

## how to call/invoke functions

```python
def show(name, gpa):
    print(name, gpa)
```

```python
# 1. simply pass values to the function as arguments, also called positional arguments
show("Chao", 4.0)
```

```python
# 2. use the parameters as keywords and then assign argument to the keywords, also called keyword-argument pairs
show(name = "Chao", gpa = 4.0)
```

```python
# 3. positional arguments should always precede keyword-argument pairs.
show("Chao", gpa = 4.0)
```

```python
# the following is incorrect
show(name = "Chao", 4.0)
```

# `lambda` expressions

```python
addone = lambda x: x+1
addone(100)
```

```python
sumup = lambda x, y: x + y
sumup(10, -6)
```

```python
compare = lambda x: x*2 if x < 10 else x
compare(5)
```

```python
# Be careful with the syntax
# this is to asign 100 to x and then execute the lambda function
v = (lambda x: x * x)(100)
v
```

# Use `lambda` expressions with `filter` and `map`

```python
# Take a list of numbers.
my_list = [12, 65, 54, 39, 102, 339, 221, 50, 70]

# use anonymous function to filter and comparing
# if divisible or not
result1 = filter(lambda x: (x % 13 == 0), my_list)

print(list(result1))
```

```python
type(result1)
```

```python
list(filter(lambda x: x > 5, range(1, 10)))
```

```python
numbers = [1, 2, 3, 4]
result2 = map(lambda x: x + x, my_list)
list(result2)
```

```python
# map two iterables to a function with two parameters
num1 = [1, 2, 3]
num2 = [4, 5, 6]

list(map(lambda x, y: x + y, num1, num2))
```

```python
print(list(map(abs, [-2, 10, -9])))
print(list(map(int, ['-2', '10', '-9'])))
print(list(map(list, ['marketing', 'finance'])))
```
