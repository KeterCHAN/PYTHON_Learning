```python
fruit = 'banana'
for letter in fruit:
    print(letter)
```

```python
range(len(fruit))
```

```python
for i in range(len(fruit)):
    print(i, fruit[i])
```

# f-strings

```python
name = "Chao"
gpa = 4.0
f"Hello, {name}. Your GPA is {gpa}."
# it's also ok to use capital F.
```

```python
# apply methods on f-strings

f"{name.upper()} is funny."
```

```python
# define multiple lines of f-strings

course = "MSBA7001"
affiliation = "HKU"
message = (
    f"Hi {name}. "
    f"You teach {course} "
    f"at {affiliation}."
)
print(message)
```

```python
print(f"'GeeksforGeeks'")

print(f"""Geeks"for"Geeks""")

print(f'''Geeks'for'Geeks''')
```

```python
greeting = "hello"
print(f"left justification: {greeting:<10}")
print(f"right justification: {greeting.replace('l', 'p'):>10}")
print(f"center justification: {greeting:^20}")
```

```python
num = 12
print(f"A width of 6 and is centered: {num:^6}")
print(f"Binary representation: {num:>15b}")
print(f"Exponentiation representation: {num:e}")
print(f"Hexidecimal representation: {num:X}")
print(f"Percentage representation: {num:%}")
print(f"Floating point representation: {num:f}")
print(f"Floating point representation with two decimal points: {num:.2f}")
# be aware of the period before 2
```
