```python
print('Welcome to')
# by default, the end of a print  is a new line.
print('HKU')
```

```python
print('Welcome to', end = ' ')
# you can change how a print ends by using end = ''
print('HKU')
```

```python
print('HKU','Business School', sep = '|', end = '@')
print('boot camp')
```
