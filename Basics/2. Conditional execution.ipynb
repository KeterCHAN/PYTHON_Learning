# `if` statement

## Simple  conditions

```python
temp = 15
if float(temp) < 20:
    print('A cold day!')
```

```python
if True:
    print('Python is such a powerful language!')
```

```python
x = 5

if x == 5: print('Equals 5')  # also ok to place them in the same line
if x > 4: print('Greater than 4')
if x >= 5: print('Greater than or Equals 5')   # fine to have spaces in the condition
if x < 6: print('Less than 6')
if x != 6: print('Not equal 6')
```

## Identation

```python
x = 5
if x > 2:
    print('Bigger than 2')
    print('Still bigger')
    print('Done with 2')
for i in range(5):
    print(i)
    if i > 2:
        print('Bigger than 2')
    print('Done with i', i)  # print is in the "for" block not "if" block
print('All Done')
```

## Chained Decisions

```python
x = 3 if 7 > 4 else -1    # although simple, but not recommended for beginners
x
```

```python
x = 9
if x < 2:
    print('Below 2')
elif x < 20:
    print('Below 20')
elif x < 10:
    print('Below 10')
else:
    print('Something else')   # what is the problem
```
